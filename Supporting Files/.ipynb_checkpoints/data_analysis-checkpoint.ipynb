# Task2_A

```python
original_time = [3.54136, 3.27827, 3.72438, 3.34107, 3.61651, 3.66029, 3.73083, 4.07393, 3.64318, 4.92446]
imrpoved_time = [2.63679, 2.31044, 2.36693, 2.67324, 2.51085, 2.50984, 2.66308, 2.69766, 2.71722, 2.54553]
```

```python
t_in_sec = [i for i in range(10)]
plt.plot(t_in_sec, original_time,'-',color='g',  label = 'Original Function')
plt.plot(t_in_sec, imrpoved_time,'-', color='orange', label = 'Improved Function')
plt.xlabel('Number of Run')
plt.ylabel('Time in seconds')
plt.legend()
plt.show()
```
